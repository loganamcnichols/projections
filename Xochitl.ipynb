```python
t = np.arange(1,1000)
z = np.arange(1,1000)
w = np.arange(1,1000)
x = 16*np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos (2 * t) - 2 * np.cos (3 * t) - np.cos(4 * t)
x2 = 16*np.sin(z)**3
y2 = 13 * np.cos(z) - 5 * np.cos (2 * w) - 2 * np.cos (3 * w) - np.cos(4 * w)
x3 = 16*np.sin(w)**3
y3 = 13 * np.cos(w) - 5 * np.cos (2 * w) - 2 * np.cos (3 * w) - np.cos(4 * w)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.plot(x, y, color='red')
ax.plot(x2/1.11, y2/1.11, color='orange')
ax.plot(x3/1.222, y3/1.222, color='yellow')
ax.set_title("For My Love Xochtil")
plt.xticks([])
plt.yticks([])
plt.show()
```
